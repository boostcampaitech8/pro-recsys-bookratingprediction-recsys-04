```python
import pandas as pd

# --------------------------------------------------
# 1) íŒŒì¼ ê²½ë¡œ
# --------------------------------------------------
path = "/data/ephemeral/home/data/features/v1/users/users_features.parquet"

# --------------------------------------------------
# 2) Parquet ì½ê¸°
# --------------------------------------------------
df = pd.read_parquet(path)

# --------------------------------------------------
# 3) ê¸°ë³¸ ì •ë³´ ì¶œë ¥
# --------------------------------------------------
print("ğŸ”¥ Data Loaded!")
print("ğŸ“Œ Shape :", df.shape)
print("\nğŸ“Œ Columns :")
print(df.columns)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------
# 1) user_review_count ê³„ì‚°
# --------------------------------------
user_review_count = (
    df.groupby("user_id")["isbn"].count().rename("user_review_count")
)

# --------------------------------------
# 2) ë¶„í¬ ì‹œê°í™” (Histogram)
# --------------------------------------
plt.figure(figsize=(12, 6))
sns.histplot(user_review_count, bins=50, kde=True)
plt.title("Distribution of user_review_count")
plt.xlabel("Number of Ratings (user_review_count)")
plt.ylabel("Number of Users")
plt.grid(True)
plt.show()

# --------------------------------------
# 3) ë¡œê·¸ ë³€í™˜ ë¶„í¬ (ì™œ qcut ì „ì— ë³´ëŠ”ê²Œ ì¢‹ìŒ)
# --------------------------------------
plt.figure(figsize=(12, 6))
sns.histplot(np.log1p(user_review_count), bins=50, kde=True)
plt.title("Distribution of log(user_review_count + 1)")
plt.xlabel("log1p(user_review_count)")
plt.ylabel("Number of Users")
plt.grid(True)
plt.show()
```
